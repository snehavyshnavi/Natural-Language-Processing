```python

import nltk
from nltk.corpus import wordnet as wn

nltk.download('wn')

word = input("Enter a word: ")


synsets = wn.synsets(word)

print(f"\nPolysemy (number of meanings): {len(synsets)}")

print("\nHomonyms (different meanings):")
for s in synsets:
    print("-", s.definition())


print("\nHyponyms (more specific words):")
for s in synsets:
    hyponyms = s.hyponyms()
    for h in hyponyms:
        print("-", h.lemmas()[0].name())
```
